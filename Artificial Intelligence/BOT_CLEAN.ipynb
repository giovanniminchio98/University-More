```python
# my idea and code, not optimized
def next_move(posr, posc, matrix):
    posx = posr
    posy = posc
    
    # i save cost and direction of each dirty cell, aim to closer to one and clean it
    # naive algorithm
    lista=[]
    
    # rows
    for i in range(len(matrix)):
        # cols
        for j in range(len(matrix[0])):
            

            if matrix[i][j] == 'd':
                # choose where to move to reash this cell
                if posx==i and posy==j:
                    print("CLEAN")
                elif posx-i < 0:
                    where = 'DOWN'
                elif posx-i > 0:
                    where='UP'
                    
                elif posy-j < 0:
                    where='RIGHT'
                elif posy-j > 0:
                    where='LEFT'

                # add dirty cell to list
                lista.add([math.sqrt((posx-i)**2 + (posy-j)**2), where])
                
    lista= lista.sort()
    
    return (str(lista[0][1])
```
